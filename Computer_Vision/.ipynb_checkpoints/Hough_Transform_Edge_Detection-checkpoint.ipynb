```python
from scipy import signal
import numpy as np
from matplotlib import pyplot as plt
import time
from scipy.ndimage.filters import gaussian_filter
from scipy.ndimage.filters import convolve
```

```python
#Helper function for plotting images
def PlotImage(filename="blocks.png",threshold=15,t='GM',first=True):
    if filename == "2D_White_Box.png":
        threshold = 40
    image = plt.imread(filename)*255
    if t == 'GM':
        i,x,y = ComputeGM(image,threshold)
    elif t == 'GO':
        i,x,y = ComputeGO(image)
    elif t == "Z":
        i,x,y = ZeroCrossings(image)
    else:
        i,x,y = FindEdges(image,threshold)
    j,k = 2,1
    if t in ['GO','GM']:
        j = 4
    fig = plt.figure(figsize=(20,20))
    plt.subplot(1,j,k)
    k+=1
    plt.title("Original")
    plt.imshow(image,cmap=plt.cm.gray)
    plt.subplot(1,j,k)
    k+=1
    plt.title("Result")
    plt.imshow(i,cmap=plt.cm.gray)
    if t in ['GO','GM']:
        plt.subplot(1,j,k)
        k+=1
        plt.title("X Direction")
        plt.imshow(x,cmap=plt.cm.gray)
        plt.subplot(1,j,k)
        k+=1
        plt.title("Y Direction")
        plt.imshow(y,cmap=plt.cm.gray)
    plt.show()
    if first:
        PlotImage("2D_White_Box.png",threshold,t,False)
```

## Edge Detection

<br><br><b>A1. Gradient Magnitude</b>

```python
# Part 1. Compute the Gradient Magnitude
def ComputeGM(img,threshold=40):
    whiteVal = 255
    gm = np.zeros_like(img)
    x = (img[1:]-img[:-1])[:,:-1]
    y = (img[:,1:]-img[:,:-1])[:-1]
    gm[:-1,:-1] = np.sqrt(x**2+y**2)
    gm[:,-1],gm[-1,:] = -1,-1
    gm[gm > threshold] = whiteVal
    gm[gm <= threshold] = 0
    return gm,y,x
            
# Show results on the blocks image
PlotImage(threshold=15)
```

<br><b>A2. Gradient Orientation</b>

```python
# Part 2. Compute the Gradient Orientation
def ComputeGO(img):
    img = img/255
    x = (img[1:]-img[:-1])[:,:-1]
    y = (img[:,1:]-img[:,:-1])[:-1]
    angle = np.arctan2(y,x)
    return angle,y,x

PlotImage(t="GO")
```

<br><b>A3. Laplacian Zero-Crossings</b> - Applying a 3x3 Laplacian kernel to blocks and find the zero crossings.



```python
# Part 3. Find the Zero Crossings
def laplacian(img,eps=0.01):
    kernal = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
    img = signal.convolve2d(img,kernal,'same')
    img[abs(img) < eps] = 1
    img[img > 1] = 0
    img[img < eps] = 0
    return img*255
def ZeroCrossings(img,showL=True):
    img = laplacian(img)
    if showL:
        plt.title("Laplacian")
        plt.imshow(img,cmap=plt.cm.gray)
        plt.show()
    imgRslt1 = np.zeros_like(img)
    imgRslt2 = np.zeros_like(img)
    sign = 1
    for j in range(len(img[0])):
        for i in range(len(img)):
            if np.sign(img[i,j]) != sign:
                imgRslt1[i,j] = 255
                sign *= -1
    sign = 1
    for i in range(len(img)):
        for j in range(len(img[0])):
            if np.sign(img[i,j]) != sign:
                imgRslt2[i,j] = 255
                sign *= -1
    imgRslt = np.hypot(imgRslt1,imgRslt2)%255
    return imgRslt,imgRslt,imgRslt

# Show the results on blocks image
PlotImage(t="Z")
```

<br><b>A4. Combine Gradient Magnitude and Zero-Crossings</b> 

```python
# Part 4. Final Solution
def FindEdges(img,threshold,alpha=0.5):
    if len(np.shape(img)) == 3:
        img = img[:,:,0]*0.2989 + img[:,:,1]*0.5870 + img[:,:,2]*0.1140
        if len(img[img > 1]) == 0 :
            img = img*255
    kernal = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
    img = gaussian_filter(img,alpha)
    res = ComputeGM(img,threshold)[0]
    plt.imshow(res,cmap=plt.cm.gray)
    plt.show()
    imgRslt = ZeroCrossings(res,showL=False)[0]
    return imgRslt,imgRslt,imgRslt


# Show the results on blocks image
start = time.time()
PlotImage(t="fZ",threshold=14)
print(time.time()-start)
```

# Hough Transform

First I processed the image by grayscaling it and using the gradient magnitude with an added gaussian blur. This helped reduce the amount of noise in the photo. Then I generated the accumulator array. I inialized it, a value equal to the circle radius I was searching for, larger than the original photo. Then I went through each pixel in the image until I found an edge (as marked by my initial processing). Once I found an edge, I used the midpoint circle algorithm to increment values in the array--The equivalent of drawing a circle in a matrix. The result of this process is seen in the displayed accumulator array images.

After generating the accumulator array, I looped through the array searching for values that were incremented beyond some threshold. When I stumbled upon one, I compared it to the average of all the pixels around it. If it was significantly higher, then I considered it the center of a circle, used the midpoint circle algorithm to draw it on my resultant image and continued on. To ensure that I got the images that were on the edges, I considered the padded portions of the image seperately. I used the same method of finding circles, but lowered the threshold for the maxima since there wouldn't be as many 'votes'.

I did this process for each size circle I was looking for. This gave me greater power to see each image since some were obscured by noise and others thrived in it. Finally, I returned my result the same size as the original image and overlayed the pictures.

If I were to work on this further, I might consider doing this for every chanel of color instead of grayscaling it.

```python
def drawAccumulatorCircle(img,i,j,radius):
    """Increments values in a circle in the matrix img with a center at i,j and a radius of radius"""
    f = 1 - radius
    ddf_x = 1
    ddf_y = -2*radius
    x = 0
    y = radius
    if j+radius < len(img[0]):
        img[i,j+radius] += 1
    if j-radius > -1:
        img[i,j-radius] += 1
    if i+radius < len(img[0]):
        img[i+radius,j] += 1
    if i-radius > -1:
        img[i-radius,j] += 1
    while x < y:
        if f >=0:
            y-=1
            ddf_y+=2
            f+=ddf_y
        x+=1
        ddf_x+=2
        f+=ddf_x
        if i+x < len(img) and j+y < len(img[0]):
            img[i+x,j+y] += 1
        if i-x > -1 and j+y < len(img[0]):
            img[i-x,j+y] += 1
        if i+x < len(img) and j-y > -1:
            img[i+x,j-y] += 1
        if i-x > -1  and j-y > -1 :
            img[i-x,j-y] += 1
        if i+y < len(img) and j+x < len(img[0]):
            img[i+y,j+x] += 1
        if i+y < len(img) and j-x > -1:
            img[i+y,j-x] += 1
        if i-y > -1 and j+x < len(img[0]):
            img[i-y,j+x] += 1
        if i-y > -1  and j-x > -1 :
            img[i-y,j-x] += 1
    return img

def CreateAccumulator(img,rad,lowBound,displayIt=False):
    """Creates an accumulator array for the image with a specified radius"""
    def pad_with(vector, pad_width, iaxis, kwargs):
        pad_value = kwargs.get('padder', 0)
        vector[:pad_width[0]] = pad_value
        vector[-pad_width[1]:] = pad_value
        return vector          #pad the image with 'rad' pixels to check for circles partially off the screen
    paddedImg = np.pad(img,rad,pad_with)
    accumulator = np.zeros_like(paddedImg)
    for i in range(len(paddedImg)):
        for j in range(len(paddedImg[0])):
            if paddedImg[i,j] >= lowBound:   #if we found an edge, draw a circle around it
                accumulator = drawAccumulatorCircle(accumulator,i,j,rad)
    if displayIt:
        plt.imshow(accumulator,cmap=plt.cm.gray)
        plt.show()
    return accumulator

def Ho(rad,accumulator,tol = 30,thresh = 20,lowBound=1.1):
    """Implements the Hough transformation for circles"""
    result = np.zeros_like(accumulator)
    iskipList = []
    jskipList = []
    #loop through the accumulator and find elements that had many intersections
    for i in range(len(accumulator)):
        for j in range(len(accumulator[0])):
            if j in jskipList and i in iskipList:                #if we find a maximum, don't look for another max close to it
                continue
            if accumulator[i,j] > thresh:
                avg = np.mean(accumulator[i-5:i+6,j-5:j+6])      #check to see if this local max is not just noise
                if accumulator[i,j] - avg > tol:
                    result += drawAccumulatorCircle(result,i,j,rad)
                    iskipList += [x for x in range(i-5,i+5)]
                    jskipList += [x for x in range(j-5,j+5)]
            elif j < rad:                                        #circles on the edge of the picture won't have as high number of intersections
                if accumulator[i,j] >= 10:
                    avg = np.mean(accumulator[i-8:i+8,j-7:j+5])  #but will have a lot more fake maxima
                    if accumulator[i,j] - avg >= tol*lowBound:
                        result += drawAccumulatorCircle(result,i,j,rad)
                        iskipList += [x for x in range(i-2,i+2)]
                        jskipList += [x for x in range(j-2,j+2)]
    #return the result the same shape as the original picture
    return result[rad:-rad,rad:-rad]

def FindEdgesCircle(img,threshold,alpha=0.5):
    """grayscales color pictures, applies a gaussian filter and finds the gradiant magnitude"""
    if len(np.shape(img)) == 3:
        img = img[:,:,0]*0.2989 + img[:,:,1]*0.5870 + img[:,:,2]*0.1140
        if len(img[img > 1]) == 0 :
            img = img*255
    kernal = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
    img = gaussian_filter(img,alpha)
    res = ComputeGM(img,threshold)[0]
    return res
```

```python
file_path = 'simplecircle.png'
oimg = plt.imread(file_path)
thresh = 1
alpha = 0.95
img = FindEdgesCircle(oimg,thresh,alpha)
rad = 64//2
lowBound=1
accsimple = CreateAccumulator(img,rad,lowBound)
l=30
fig = plt.figure(frameon=False,figsize=(8,8))
plt.subplot(121)
plt.imshow(oimg)
imagesimple = Ho(rad,accsimple,tol=24,thresh=l,lowBound=1.1)
temp = np.zeros_like(oimg)
temp[:,:,1]=imagesimple
plt.imshow(temp,alpha=0.4)
plt.subplot(122)
plt.imshow(accsimple,cmap=plt.cm.gray)
plt.show()
```

```python
file_path = 'circles.png'                            #get the original picture so we can process it
oimg = plt.imread(file_path)
thresh = 18                                          #initialize useful variables that we might change
alpha = 0.95
img = FindEdgesCircle(oimg,thresh,alpha)             #Calculate the edges
rad = 32//2
acc1 = CreateAccumulator(img,rad,1)                  #Create the accumulator picture for diameter 32
image1 = Ho(rad,acc1,tol=24,thresh=30,lowBound=1.1)  #use the Hough transformation to find the circles
image1[image1 != 0]=1
fig = plt.figure(figsize=(15,15))
plt.subplot(121)
plt.imshow(oimg)
temp = np.zeros_like(oimg)
temp[:,:,1]+=image1
temp[:,:,0]+=image1
temp[:,:,2]+=image1
plt.imshow(temp,alpha=0.4)
plt.title("Detected Circles")
plt.subplot(122)
plt.title("Accumulator Array for Diameter {}".format(32))
plt.imshow(acc1,cmap=plt.cm.gray)
plt.show()
```
